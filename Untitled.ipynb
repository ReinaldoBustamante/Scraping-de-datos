```python
def read_shapefile(sf):
    """
    Read a shapefile into a Pandas dataframe with a 'coords' 
    column holding the geometry information. This uses the pyshp
    package
    """
    fields = [x[0] for x in sf.fields][1:]
    records = sf.records()
    shps = [s.points for s in sf.shapes()]
    df = pd.DataFrame(columns=fields, data=records)
    df = df.assign(coords=shps)
    return df
```

#### Funciones utiles

```python
def plot_regiones_data(sf, title, regiones, data=None, 
                      color=None, print_id=False):
    '''
    Plot map with selected comunes, using specific color
    '''
    
    color_ton, bins = calc_color(data, color)
    df = read_shapefile(sf)
    region_id = []
    for i in regiones:
        #print(i)
        id = df[df.NOMBRE == i].index.get_values()[0]
        region_id.append(id)
        
        
    y_lim = (-41,-17) # latitude 
    x_lim = (-76, -66) # longitude
    
    plot_map_fill_multiples_ids_tone(sf, title, region_id, 
                                     print_id, 
                                     color_ton, 
                                     bins, 
                                     x_lim = x_lim, 
                                     y_lim = y_lim, 
                                     figsize = (10,25));
```

```python
def calc_color(data, color=None):
        if color   == 1: 
            color_sq =  ['#dadaebFF','#bcbddcF0','#9e9ac8F0',
                        '#807dbaF0','#6a51a3F0','#54278fF0']; 
            colors = 'Purples';
        elif color == 2: 
            color_sq = ['#c7e9b4','#7fcdbb','#41b6c4',
                        '#1d91c0','#225ea8','#253494']; 
            colors = 'YlGnBu';
        elif color == 3: 
            color_sq = ['#f7f7f7','#d9d9d9','#bdbdbd',
                        '#969696','#636363','#252525']; 
            colors = 'Greys';
        elif color == 9: 
            color_sq = ['#ff0000','#ff0000','#ff0000',
                        '#ff0000','#ff0000','#ff0000']
        else:            
            color_sq = ['#ffffd4','#fee391','#fec44f',
                        '#fe9929','#d95f0e','#993404']; 
            colors = 'YlOrBr';
        
        new_data, bins = pd.qcut(data, 6, retbins=True, labels=list(range(6)))
        color_ton = []
        for val in new_data:
            color_ton.append(color_sq[val]) 
        if color != 9:
            colors = sns.color_palette(colors, n_colors=6)
            sns.palplot(colors, 0.6);
            for i in range(6):
                print ("\n"+str(i+1)+': '+str(int(bins[i]))+
                       " => "+str(int(bins[i+1])-1), end =" ")
            print("\n\n   1   2   3   4   5   6")    
        return color_ton, bins;
```

```python
def plot_map_fill_multiples_ids_tone(sf, title, comuna,  
                                     print_id, color_ton, 
                                     bins, 
                                     x_lim = None, 
                                     y_lim = None, 
                                     figsize = (5,12)):
    '''
    Plot map with lim coordinates
    '''
        
    plt.figure(figsize = figsize)
    fig, ax = plt.subplots(figsize = figsize)
    fig.suptitle(title, fontsize=16)
    for shape in sf.shapeRecords():
        x = [i[0] for i in shape.shape.points[:]]
        y = [i[1] for i in shape.shape.points[:]]
        ax.plot(x, y, 'k')
            
    for id in comuna:
        shape_ex = sf.shape(id)
        x_lon = np.zeros((len(shape_ex.points),1))
        y_lat = np.zeros((len(shape_ex.points),1))
        for ip in range(len(shape_ex.points)):
            x_lon[ip] = shape_ex.points[ip][0]
            y_lat[ip] = shape_ex.points[ip][1]
        ax.fill(x_lon,y_lat, color_ton[comuna.index(id)])
        if print_id != False:
            x0 = np.mean(x_lon)
            y0 = np.mean(y_lat)
            plt.text(x0, y0, id, fontsize=10)
    if (x_lim != None) & (y_lim != None):     
        plt.xlim(x_lim)
        plt.ylim(y_lim)

```

### Pregunta 8: Implementar un script Python para mostrar las 10 palabras claves ("mas frecuentes") de cada categoria de anuncios.

```python
#!pip install collections
#!pip install spacy
#!python -m spacy download es_core_news_sm
```

```python
import spacy 
from collections import Counter
import pandas as pd
import mysql.connector
db_connection = mysql.connector.connect(user="root",host="localhost",password="spawn1997")
cursor = db_connection.cursor()
cursor.execute("use Yapo")
nlp = spacy.load("es_core_news_sm")
spacy_stopwords = spacy.lang.es.stop_words.STOP_WORDS
pd.options.display.max_colwidth = 300
```

```python
#se concatena todas las descripciones por categoria con group_concat
sql = "select categoria, GROUP_CONCAT( DISTINCT descripcion ORDER BY descripcion SEPARATOR', ') as concat_description from anuncio group by categoria;"
cursor.execute(sql)
df1 = pd.DataFrame(cursor.fetchall())
df1.columns = cursor.column_names
#dejaremos como indice del dataFrame a la columna categoria
df1.index = df1.categoria 
df1.drop(['categoria'], axis='columns',inplace = True)
df1
```

```python
palabras_claves = []
for i in range(df1.index.size):
    doc = nlp(df1.iloc[i].concat_description)
    #obtendremos las palabras de doc que sean adjetivos,sustantivos y verbos
    palabras = [token.text for token in doc if token.is_stop != True and token.is_punct != True and (token.pos_ == "ADJ" or token.pos_ == "VERB" or token.pos_ == "NOUN")]
    #obtendremos la frecuencia de cada palabra
    frecuencia = Counter(palabras)
    #obtenemos las 10 mas comunes
    palabras_mas_frecuentes = frecuencia.most_common(10)
    #las agregamos al arreglo creado inicialmente
    palabras_claves += [palabras_mas_frecuentes]
```

```python
#creacion del segundo dataframe con las 10 palabras_claves mas frecuentes de tipo [(palabra,frecuencia),..........]
df2 = pd.DataFrame({'categoria': df1.index, 'Palabras claves' : palabras_claves})
df2.index = df2.categoria
df2.drop(['categoria'], axis = 1, inplace = True)
```

```python
sql = "select vendedor.nombre_region as region, GROUP_CONCAT( DISTINCT descripcion SEPARATOR', ' ) as concat_description from anuncio JOIN vendedor using(id_vendedor) group by vendedor.nombre_region;" 
cursor.execute(sql)
df1 = pd.DataFrame(cursor.fetchall())
df1.columns = cursor.column_names
#dejaremos como indice del dataFrame a la columna categoria
df1.index = df1.region
df1.drop(['region'], axis='columns',inplace = True)
df1
```

```python
palabra_clave = []
for i in range(df1.index.size):
    doc = nlp(df1.iloc[i].concat_description)
    #obtendremos las palabras de doc que sean adjetivos,sustantivos y verbos
    palabras = [token.text for token in doc if token.is_stop != True and token.is_punct != True and (token.pos_ == "ADJ" or token.pos_ == "VERB" or token.pos_ == "NOUN")]
    #obtendremos la frecuencia de cada palabra
    frecuencia = Counter(palabras)
    #obtenemos las 10 mas comunes
    palabras_mas_frecuentes = frecuencia.most_common(1)
    #las agregamos al arreglo creado inicialmente
    palabra_clave += [palabras_mas_frecuentes]
```
